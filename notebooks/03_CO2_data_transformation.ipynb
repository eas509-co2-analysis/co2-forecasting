In this notebook we will check whether log transforming the `CO2 (ppm)` column help in making it more Gaussian.

# Imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# Reading the data

```python
DAILY_DATA_PATH = "../data/interim/daily_data.csv"
df = pd.read_csv(DAILY_DATA_PATH)
df["Date Time"] = pd.to_datetime(df["Date Time"])
df.set_index("Date Time", inplace=True)
df.head()
```

# Does log transforming the CO2 column help?

```python
sns.histplot(df['CO2 (ppm)'], kde=True);
```

```python
sns.histplot(np.log(df['CO2 (ppm)']), kde=True);
```

No the distribution has more or less the same shape. It is already close to normal before transformation.
