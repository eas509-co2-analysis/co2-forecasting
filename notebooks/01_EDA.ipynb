# Duplicates

```python
df[df.index.duplicated(keep=False)].sort_index()
```

# Investigating the Temporal Resolution

The data has a temporal resolution of 10 minutes, which is extremely high. This will result in huge noise in the data. So, we will check if downscaling this temporal resolution to daily, weekly, and monthly makes the data more manageable.

## Daily vs. Weekly Resolution

```python
df_daily = df_with_nan.resample("D").mean()
df_daily
```

```python
for variable in df_daily.columns:
    df_daily[variable].plot(figsize=(15,5), title=f"Daily {variable} Over Time")
    plt.ylabel(f"{variable}")
    plt.grid(True)
    plt.show();
```

```python
df_weekly = df_with_nan.resample("W").mean()
df_weekly
```

```python
for variable in df_weekly.columns:
    df_weekly[variable].plot(figsize=(15,5), title=f"Weekly {variable} Over Time")
    plt.ylabel(f"{variable}")
    plt.grid(True)
    plt.show();
```

**Observations:**
* Daily resolution includes sufficiently large number of data points and also sufficient complexity in terms of the seasonality and the trend component as compared to the weekly resolution. Hence, we will go with daily resolution.
